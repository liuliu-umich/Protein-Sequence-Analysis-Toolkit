```python
#import onehot and esm representation of sequences
df = pd.read_csv("Datasets/VDJdb_PosNeg_corrected_onehot_esm_sample.csv")

onehot_cols_to_rename = [col for col in df.columns if col.startswith('epitope_embedding') or col.startswith('cdr3_embedding')]
df = df.rename(columns={col: 'onehot_' + col for col in onehot_cols_to_rename})

esm_cols_to_rename = [col for col in df.columns if col.startswith('epitope_emb_') or col.startswith('cdr3_emb_')]
df = df.rename(columns={col: 'esm_' + col for col in esm_cols_to_rename})

labels = df["Binding"]
```

```python
X_onehot = protein_data.extract_columns_by_prefix('onehot_')[0]
X_esm = protein_data.extract_columns_by_prefix('esm_')[0]
labels = protein_data.extract_columns_by_prefix('Binding')[0]
```

# Dimensionality Reduction

```python
onehot_analyzer = PCAAnalyzer(n_components=10)

onehot_component_candidates = list(range(1, X_onehot.shape[1] + 1))
onehot_best_components = onehot_analyzer.tune_components(X_onehot, onehot_component_candidates)
print(f"Optimal number of components: {onehot_best_components}")

# Fit and transform data
onehot_analyzer.n_components = onehot_best_components
pca_onehot_features = onehot_analyzer.fit_transform(X_onehot)

# Plot the explained variance
onehot_analyzer.plot_variance()

# Plot the 2D PCA projection with labels
onehot_analyzer.plot_2d_projection(pca_onehot_features, labels=labels.values)

# Save the model
onehot_analyzer.save_model("pca_model.joblib")
```

```python
esm_analyzer = PCAAnalyzer(n_components=10)

esm_component_candidates = list(range(1, X_esm.shape[1] + 1))
esm_best_components = esm_analyzer.tune_components(X_esm, esm_component_candidates)
print(f"Optimal number of components: {esm_best_components}")

# Fit and transform data
esm_analyzer.n_components = esm_best_components
pca_esm_features = esm_analyzer.fit_transform(X_esm)

# Plot the explained variance
esm_analyzer.plot_variance()

# Plot the 2D PCA projection with labels
esm_analyzer.plot_2d_projection(pca_esm_features, labels=labels.values)

# Save the model
esm_analyzer.save_model("esm_model.joblib")
```
